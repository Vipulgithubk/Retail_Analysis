```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Data = pd.read_csv("/content/SampleSuperstore (1).csv")
Data.head(10)
```

```python
Data.tail(10)
```

```python
print(Data.describe())
```

```python
print(Data.isnull().sum())
```

```python
print(Data['Category'].unique())
```

```python
plt.figure(figsize=(12, 10))
sns.heatmap(Data.corr(), annot=True)
plt.show()
```

```python
plt.scatter(Data['Sales'], Data['Profit'])
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.show()
```

```python
plt.bar(Data['Category'], Data['Sales'])
plt.xlabel('Category')
plt.ylabel('Sales')
plt.show()
```

```python
sns.stripplot(x="Region",y="Sales",data=Data,jitter=0.2,hue="Category")
```

```python
sns.countplot(x="Region",data=Data)
```

```python
sns.countplot(x="Region",data=Data,hue="Category")
```

```python
sns.countplot(x="Category",data=Data,hue="Region")
```

```python
sns.countplot(x="Quantity",data=Data,hue="Category")
```

```python
sns.stripplot(x="Region",y="Quantity",data=Data,jitter=0.2,palette="Set1")
```

```python
sns.set(style="whitegrid")
sns.violinplot(x="Category",y="Profit",data=Data,hue="Region")
```

```python
sns.boxplot(data=Data,x="Region",y="Sales",palette="cubehelix")
```
